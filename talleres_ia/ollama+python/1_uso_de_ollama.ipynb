```python
import ollama
```

```python
modelos = ollama.list()
```

```python
print("Modelos instalados: ")
for modelo in modelos['models']:
    print(modelo['model'])
```

```python
OLLAMA_MODEL= "mistral:latest"
```

```python
response = ollama.chat(
    model=OLLAMA_MODEL, 
    messages=[
       {
           'role': 'system', 
           'content': 'eres una profe experta en analisis de datos, paciente'
       }, 
       { 
           'role': 'user',
           'content': 'explicame que es un ETL'
       } 
    ]
)
```
