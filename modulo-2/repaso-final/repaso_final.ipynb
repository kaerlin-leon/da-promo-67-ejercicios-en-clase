```python
import requests
import pandas as pd
import mysql.connector
from mysql.connector import Error
import numpy as np
import os
from dotenv import load_dotenv

load_dotenv()
```

```python
# fase 1. Extracción de datos de películas navideñas desde un API. 
## Buenas practicas: uso de try/except, uso de funciones , verificar los tipos de respuestas
url = "https://beta.adalab.es/resources/apis/christmas.json"
datos = requests.get(url)
print(datos.status_code)
print(datos.content)

datos_pelis = datos.json()
print(datos_pelis)

```

```python
#fase 2: Creación de la Base de Datos
# 2.1 conectarme a mysql 
#mejoras: contraseñas con dotenv, definir funciones
try:
    cnx = mysql.connector.connect(
        host='127.0.0.1',
        user='root',
        password='',
    )
    print('Conexión exitosa')
except Error as e:
    print('Error al conectar:', e)
```

```python
# 2.2 Crear la bases de datos
try:
    cursor = cnx.cursor()
    query = "CREATE DATABASE IF NOT EXISTS christmas" 
    cursor.execute(query)
    print("Query exitosa")
except Error as e:
    print(e) 
```

```python
cursor = cnx.cursor()
```

```python
## crear la estructura de la bases de datos: tengo que pensar si todo va en la misma tabla
# crear tabla
cursor.execute("USE christmas")

query = '''CREATE TABLE christmas_movies (
    id INT PRIMARY KEY AUTO_INCREMENT,
    title VARCHAR(255) NOT NULL,
    rating VARCHAR(20),
    runtime INT,
    imdb_rating DECIMAL(3,1),
    meta_score INT,
    genre VARCHAR(255),
    release_year INT,
    description TEXT,
    director VARCHAR(255),
    stars TEXT,
    votes VARCHAR(50),
    gross VARCHAR(50),
    img_src VARCHAR(500),
    type VARCHAR(50)
);'''

cursor.execute(query)
```

```python
df_christmas.isnull().sum()
```

```python
# insertar en la bases de datos
query_insert = '''   INSERT INTO christmas_movies (title, rating, runtime, imdb_rating, meta_score, 
                              genre, release_year, description, director, stars, 
                              votes, gross, img_src, type) 
VALUES (%s, %s, %s, %s, %s, %s, %s, %s, %s, %s, %s, %s, %s, %s)  '''

df_clean = df_christmas.replace({np.nan: None, 'nan': None, 'Nan': None})
datos = df_clean.values.tolist()
cursor.executemany(query_insert, datos)
cnx.commit()
print(f"Se han insertado nuevos {cursor.rowcount} valores")

```

```python
## consultas sobre la bases de datos 
query ='''select * 
from christmas_movies
order by imdb_rating desc
limit 10 '''

df_top = pd.read_sql(query, cnx)
df_top
```

```python
query ='''select
	floor(release_year/10) * 10 as decada, 
    count(*) as total_pelis, 
    round(avg(imdb_rating), 2) as rating_promedio
from christmas_movies
group by decada
order by decada; '''
df_decadas = pd.read_sql(query, cnx)
df_decadas
```
