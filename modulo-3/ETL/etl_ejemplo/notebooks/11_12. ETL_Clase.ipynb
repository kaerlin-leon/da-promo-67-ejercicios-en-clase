# Proceso ETL

Un proceso ETL (Extract, Transform, Load) es un flujo de trabajo que se utiliza en el análisis de datos para obtener datos de diversas fuentes, procesarlos y limpiarlos, y cargarlos en un destino final como una base de datos o un dashboard.

El término ETL viene de las siglas en inglés:

- Extract (Extraer): Obtener los datos desde distintas fuentes (bases de datos, APIs, archivos CSV, etc.).
- Transform (Transformar): Limpiar y preparar los datos para que sean útiles.
- Load (Cargar): Mover los datos procesados a un destino final donde serán usados para análisis.


## Proceso ETL (Extract, Transform, Load)

|  **ETL**        |  **Acción**                                 |  **Ejemplo práctico**                          |
|-----------------|----------------------------------------------|-------------------------------------------------|
| **E: Extract**  | Obtener datos de diferentes fuentes (CSV, API, SQL) | Descargar un archivo CSV con datos de ventas.   |
| **T: Transform** | Limpiar y transformar los datos              | Eliminar valores nulos, calcular promedios, convertir formatos de fecha. |
| **L: Load**     | Cargar los datos en un destino final          | Guardar los datos limpios en una base de datos o enviarlos a un dashboard. |

2. Crear funciones con las operaciones anteriores y llamarlas

```python
import pandas as pd
import pymysql
from sqlalchemy import create_engine

# Configuración de la base de datos
host = '127.0.0.1'
user = 'root'
password = 'AlumnaAdalab'
database = 'bd_etl'


# =====================
# EXTRACT: Obtener los datos desde los archivos CSV
# =====================
def extract_data(file_path):
    print(f"Extrayendo datos desde {file_path}...")
    data = pd.read_csv(file_path)
    print(data.head())
    return data

# =====================
# TRANSFORM: Limpiar y transformar los datos
# =====================
def transform_data(data):
    print("Transformando datos...")
    data = data.dropna()  # Eliminar valores nulos
    data.columns = data.columns.str.lower()  # Renombrar las columnas con minúsculas
    print(data.head())
    return data

# =====================
# LOAD: Crear la base de datos y cargar los datos
# =====================
def create_db():
    # Conectar a MySQL usando pymysql
    connection = pymysql.connect(
        host=host,
        user=user,
        password=password
    )

    # Crear un cursor
    cursor = connection.cursor()

    # Crear una base de datos si no existe
    cursor.execute(f"CREATE DATABASE IF NOT EXISTS {database}")
    print("Base de Datos creada exitosamente.")

    # Cerrar la conexión
    connection.close()

def load_data(table_name, data):
    print(f"Cargando datos en la tabla {table_name}...")

    # Crear conexión a MySQL usando SQLAlchemy
    engine = create_engine(f'mysql+pymysql://{user}:{password}@{host}/{database}')

    # Insertar datos desde el DataFrame en MySQL
    data.to_sql(table_name, con=engine, if_exists='append', index=False)
    print(f"Datos insertados en la tabla {table_name} exitosamente.")

# =====================
# Proceso ETL completo
# =====================
def etl_process():
    # Crear la base de datos
    create_db()

    # Extraer, transformar y cargar datos de clientes
    clientes = extract_data('clientes.csv')
    clientes_transformed = transform_data(clientes)
    load_data('clientes', clientes_transformed)

    # Extraer, transformar y cargar datos de ventas
    ventas = extract_data('ventas.csv')
    ventas_transformed = transform_data(ventas)
    load_data('ventas', ventas_transformed)

# Ejecutar el proceso ETL completo
etl_process()
```
