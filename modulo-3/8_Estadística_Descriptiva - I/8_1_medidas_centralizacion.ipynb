## Estadistica descriptiva 
- describir y resumir datos
- identificar valores atipicos
- exploracion relacion entre variables

```python
import pandas as pd
import numpy as np 
import seaborn as sns
import matplotlib.pyplot as plt
import warnings

pd.set_option('display.max_columns', None)
warnings.filterwarnings('ignore')
```

```python
df = pd.read_csv('files/bank-additional_clean.csv', index_col=0)
df
```

```python
## income
media_edad = df['age'].mean()
print(np.round(media_edad, 2))

mediana_edad = df['age'].median()
print(media_edad)

moda_edad = df['age'].mode()[0]
print(media_edad)
```

- Los ingresos medio es 93241.2 
- La mediana de los ingresos es 93050.5, que el 50% de los personas ganan menos 93050.5 y el otro 50 % gana mas que 93050.5. 
- El ingreso que mas se repite es 40655. 
- No hay una diferencia signifivativa entre la media y la mediana, por lo que podria decirse que los datos estan noramlizados,no hay gran datos extremos. 

```python
sns.histplot(x='age', data=df, bins=10)
# Visualización: Histograma con la media y la mediana
plt.axvline(media_edad, color='red', linestyle='dashed', linewidth=2, label=f'Media: {media_edad}')
plt.axvline(mediana_edad, color='green', linestyle='dashed', linewidth=2, label=f'Mediana: {media_edad}')
plt.axvline(moda_edad, color='blue', linestyle='dashed', linewidth=2, label=f'Moda: {moda_edad}')

plt.title('Distribución, Media y Mediana')
plt.legend()
plt.show()
```

```python
## income
media_ingresos = df['income'].mean()
print(np.round(media_ingresos, 2))

mediana_ingresos = df['income'].median()
print(mediana_ingresos)

moda_ingresos = df['income'].mode()[0]
print(moda_ingresos)
```

```python
sns.histplot(x='income', data=df, bins=30)
# Visualización: Histograma con la media y la mediana
plt.axvline(media_ingresos, color='red', linestyle='dashed', linewidth=2, label=f'Media: {media_ingresos}')
plt.axvline(mediana_ingresos, color='green', linestyle='dashed', linewidth=2, label=f'Mediana: {mediana_ingresos}')
plt.axvline(moda_ingresos, color='blue', linestyle='dashed', linewidth=2, label=f'Moda: {moda_ingresos}')

plt.title('Distribución, Media y Mediana')
plt.legend()
plt.show()
```

```python

plt.figure(figsize=(10, 6))
sns.boxplot(x=df['income'])

# Dibujar líneas para la media, mediana y moda
plt.axvline(media_ingresos, color='red', linestyle='dashed', linewidth=2, label=f'Media: {media_ingresos}')
plt.axvline(mediana_ingresos, color='green', linestyle='dashed', linewidth=2, label=f'Mediana: {mediana_ingresos}')
plt.axvline(moda_ingresos, color='blue', linestyle='dashed', linewidth=2, label=f'Moda: {moda_ingresos}')

# Añadir leyenda y título
plt.legend()
plt.title('Boxplot con Media, Mediana y Moda')

# Mostrar el gráfico
plt.show()
```
