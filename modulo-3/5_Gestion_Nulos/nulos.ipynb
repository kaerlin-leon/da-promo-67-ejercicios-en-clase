## Analisis de las numericas

```python
# age -->  11.9
# cons.price.idx --> 1.09
# euribor --21,52
```

```python
# Para saber de manera numerica como estám distribuidos los datos
# Si media ≈ mediana → simétrica
# Si media >> mediana → asimétrica derecha
# Si media << mediana → asimétrica izquierda
```

```python
mediana = df['age'].median()
mediana
```

```python
media = df['age'].mean()
media
```

```python
df['age'] = df['age'].fillna(mediana)
```

```python
cols_numericas = ["euribor3m", "age", "duration"]
df[cols_numericas] = IterativeImputer(max_iter=10, random_state=42).fit_transform(df[cols_numericas])
```

```python
simple_imputer = SimpleImputer(strategy="most_frequent")
df["education"] = simple_imputer.fit_transform(df[['education']]).ravel() #fit_transform devuelve un array 2D de NumPy, y lo tenemos que asignar a una columa
```

```python
simple_imputer.fit_transform(df[['education']])
```

| Estrategia | Descripción | Tipo de dato |
|---|---|---|
| `"mean"` | Reemplaza con la **media** de la columna | Numérico |
| `"median"` | Reemplaza con la **mediana** de la columna | Numérico |
| `"most_frequent"` | Reemplaza con la **moda** (valor más frecuente) | Numérico y categórico |
| `"constant"` | Reemplaza con un **valor fijo** definido por `fill_value` | Numérico y categórico |

```python
# euribor --21,52
knn_imputer = KNNImputer(n_neighbors=4)
cols_numericas = ["euribor3m", "age", "duration"]  # las que tengas
df[cols_numericas] = knn_imputer.fit_transform(df[cols_numericas])

```
