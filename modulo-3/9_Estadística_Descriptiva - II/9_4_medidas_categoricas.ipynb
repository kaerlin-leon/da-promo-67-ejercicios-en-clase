```python
import pandas as pd
import numpy as np 
import seaborn as sns
import matplotlib.pyplot as plt
import warnings

pd.set_option('display.max_columns', None)
warnings.filterwarnings('ignore')
```

```python
df = pd.read_csv('files/bank-additional_clean.csv', index_col=0)
df.head(3)
```

```python
## frecuencia absoluta 
# frecuencia relativa
```

```python
frecuencia_absoluta = df['marital'].value_counts()
frecuencia_absoluta.reset_index

frecuencia_relativa = df['marital'].value_counts(normalize=True)*100
frecuencia_relativa

tabla_frecuencias = pd.DataFrame({
    'estado civil': frecuencia_absoluta.index, 
    'frecuencia_absoluta': frecuencia_absoluta.values,
    'frecuencia_relativa': frecuencia_relativa.values
}
)
tabla_frecuencias
```

```python
## grafico barras = countplot

sns.countplot(x='marital', data=df)
```

```python
## relacion entre dos variables categoricas
orden_education = df['education'].value_counts().index
sns.countplot(x='marital', data=df, hue='education', hue_order=orden_education)
```

```python
## crosstab

tabla_contingencia = pd.crosstab(df['marital'], df['education'], normalize=True)*100
tabla_contingencia
```
