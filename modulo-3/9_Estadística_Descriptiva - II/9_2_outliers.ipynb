```python
df = pd.read_csv('files/bank-additional_clean.csv', index_col=0)
df.head(3)
```

# Outliers: 
- un valor que sale de lo normal. 

```python
q1 = df['age'].quantile(0.25)
q3 = df['age'].quantile(0.75)

iqr = q3 - q1
```

```python
lower = q1 - 1.5*iqr
upper = q3 + 1.5*iqr
```

```python
outliers = df[(df['age']< lower) | (df['age']> upper)]
```
